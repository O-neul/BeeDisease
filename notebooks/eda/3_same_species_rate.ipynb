```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def similarity_distribution(result_df, title="Same Species Rate Distribution", save_path=None):
    """
    'Same_Species_Rate' 분포를 히스토그램 + 밀도곡선으로 시각화하고,
    save_path가 주어지면 파일로 저장합니다.
    
    Args:
        result_df (pd.DataFrame): 분석 결과 DataFrame
        title (str): 그래프 제목
        save_path (str or None): 저장 경로 (예: "output.png")
    """
    plt.figure(figsize=(10, 6))
    sns.histplot(result_df['Same_Species_Rate'].dropna(), bins=20, kde=True)
    
    plt.title(title, fontsize=14)
    plt.xlabel("Same Species Rate", fontsize=12)
    plt.ylabel("Count", fontsize=12)
    plt.grid(True)
    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')  # 고해상도 저장
        print(f"그래프 저장: {save_path}")


    plt.show()
```

```python

result = pd.read_csv("../../data/final/7_nearby_similar_cer.csv")
result1 = pd.read_csv("../../data/final/7_nearby_similar_mel.csv")
```

```python
cer = similarity_distribution(result, title="Cerana Same Species Rate distribution", save_path="../../outputs/figures/3_1_same_species_rate_cer.png")
mel = similarity_distribution(result1, title="Mellifera Same Species Rate distribution", save_path="../../outputs/figures/3_2_same_species_rate_mel.png")
```
