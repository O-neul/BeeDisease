꿀벌 종 별로 밀원의 특정 '종'이 얼마나 자주 등장했는지를 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
def species_counts(df, title="종별 빈도 분포", save_path=None):
    """
    DataFrame에서 '종' 컬럼 기준으로 종별 빈도 분포 시각화

    Args:
        df (pd.DataFrame): '종' 열이 포함된 데이터프레임
        title (str): 그래프 제목
        save_path (str): 저장할 파일 경로 (예: "species_counts.png"), None이면 저장하지 않음
    """
    species_counts = df['종'].value_counts()

    plt.figure(figsize=(10, 6))
    sns.barplot(x=species_counts.values, y=species_counts.index, orient='h')
    
    plt.xlabel("Count", fontsize=12)
    plt.ylabel("Species", fontsize=12)
    plt.title(title, fontsize=14)
    plt.grid(True, axis='x')
    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"그래프 저장: {save_path}")
    
    plt.show()
```

```python
df = pd.read_csv("../../data/final/4_Apis_cerana.csv")
df1 = pd.read_csv("../../data/final/4_Apis_mellifera.csv")
```

```python
species_counts(df, title="재래종 밀원별 분포", save_path="../../outputs/figures/4_1_cerana_species_counts.png")
species_counts(df1, title="서양종 밀원별 분포", save_path="../../outputs/figures/4_2_mellifera_species_counts.png")
```
